```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/Week_1.csv")
df.head()
```

```python
mean_marks = df["Marks"].mean()
median_marks = df["Marks"].median()
mode_marks = df["Marks"].mode()[0]
std_dev_marks = df["Marks"].std()

mean_marks, median_marks, mode_marks, std_dev_marks
```

```python
df["Marks"].describe()
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.hist(df["Marks"], bins=6)
plt.xlabel("Marks")
plt.ylabel("Number of Students")
plt.title("Distribution of Student Marks")
plt.show()
```

```python
plt.figure()
plt.boxplot(df["Marks"], vert=False)
plt.xlabel("Marks")
plt.title("Boxplot of Student Marks")
plt.show()
```

```python
Q1 = df["Marks"].quantile(0.25)
Q3 = df["Marks"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df["Marks"] < lower_bound) | (df["Marks"] > upper_bound)]
outliers
```
